```python
import datetime
listtime=[]

for x in data['month']:
  
  monthinteger = x
  listtime.append(datetime.date(1900,monthinteger,1).strftime('%B'))
data['monthfull']=listtime
data['year_month']=[d.split('/')[2] + " " +d.split('/')[0] for d in data.date]
data
```

```python
weight_list = [] # create blank list

# loop through each hour
for x in data['Month_Year'].unique(): 
   weight_list.append(data.loc[data['Month_Year'] == x,['lat','lng','population']].groupby(['lat','lng'])\
    .sum().reset_index().values.tolist())
# preview output
weight_list
```

```python
df_date_list = [] # create blank list

# loop through each hour
for date in data['date'].sort_values().unique(): 
    df_date_list.append(data.loc[data['date'] == date,['lat','lng','population']].groupby(['lat','lng'])\
    .sum().reset_index().values.tolist())
# preview output
df_date_list
```
